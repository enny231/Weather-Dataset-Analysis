```python
sns.pairplot(data=df)
```

```python
numerical_features = ['Temperature', 'Humidity', 'Wind Speed', 'Precipitation (%)', 'Atmospheric Pressure', 'UV Index', 'Visibility (km)']
for feature in numerical_features:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'{feature} Distribution')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[feature])
    plt.title(f'{feature} Boxplot')
```

```python
categorical_features = ['Cloud Cover', 'Season', 'Location', 'Weather Type']
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=feature, order=df[feature].value_counts().index)
    plt.title(f'{feature} Distribution')
    plt.xticks(rotation=45)
    plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of Weather Types across Seasons
df.head()
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Season', hue='Weather Type')
plt.title('Weather Type Distribution Across Seasons')
plt.show()
```

```python
# Distribution of Weather Types across Locations
df.head()
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Location', hue='Weather Type')
plt.title('Weather Type Distribution Across Locations')
plt.show()
```

```python
# Temperature trends over Seasons
df.head()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Season', y='Temperature')
plt.title('Temperature Trends Across Seasons')
plt.show()
```

```python
# Temperature trends over Locations
df.head()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Location', y='Temperature')
plt.title('Temperature Trends Across Locations')
plt.show()
```

```python
# Scatter plot of Humidity vs Precipitation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Humidity', y='Precipitation (%)')
plt.title('Humidity vs Precipitation')
plt.show()
```

```python
# UV Index variations with Cloud Cover
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Cloud Cover', y='UV Index')
plt.title('UV Index Across Cloud Cover Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line plot of Visibility vs Atmospheric Pressure
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Atmospheric Pressure', y='Visibility (km)')
plt.title('Visibility vs Atmospheric Pressure')
plt.show()
```

```python
# Wind Speed across Weather Types
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Weather Type', y='Wind Speed')
plt.title('Wind Speed Across Weather Types')
plt.xticks(rotation=45)
plt.show()
```
